```python
from bs4 import BeautifulSoup 
import requests 
from csv import writer 
```

```python
#accessing the pararius website url 

url='https://www.pararius.com/apartments/amsterdam'
page=requests.get(url)
```

```python
print(page)
```

```python
#creating object 
soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup)
```

```python
lists=soup.find_all('section',class_="listing-search-item")
```

```python
print(lists)
```

```python
# creating the content 
with open('Amsterdam1.csv','w', newline='') as f:
    thewriter=writer(f)
    header=['Title','Location','price','Area','Rooms']
    thewriter.writerow(header)
    for list in lists:
        
        title= list.find('h2',class_="listing-search-item__title").text.replace('\n','')
        location=list.find('div',class_="listing-search-item__sub-title").text.replace('\n','')
        price=list.find('div',class_="listing-search-item__price").text.replace('\n','')
        area=list.find('li',class_="illustrated-features__item").text.replace('\n','')
        rooms=list.find('li',class_="illustrated-features__item illustrated-features__item--number-of-rooms").text.replace('\n','')
        info=[title,location,price,area,rooms]  
        thewriter.writerow(info)            
```
